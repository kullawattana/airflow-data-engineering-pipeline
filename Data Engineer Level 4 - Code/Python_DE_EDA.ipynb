# ⏩ Exploratory Data Analysis
Exploratory Data Analysis หรือ EDA คือ กระบวนการ ตรวจสอบ สำรวจข้อมูลเบื้องต้น เป็นการวิเคราะห์ข้อมูลที่จำเป็นก่อนการนำข้อมูลไปใช้ หรือนำไปวิเคราะห์เชิงลึก โดยประโยชน์ของการทำ EDA จะช่วยทำให้เราเข้าใจพื้นฐานเกี่ยวกับข้อมูลชุดนั้น และเป็นการตรวจความผิดพลาดของชุดข้อมูล


----------------------------------------------------------------------------------

## ⚡ สำรวจข้อมูลตาม 6 ขั้นตอน
1. Distinguish Attributes  การเลือกข้อมูล และแสดงทำเป็น คอลัมภ์
2. Univariate Analysis การวิเคราะห์ข้อมูลตัวแปรเดียว เชิงสถิติ
3. Bi-/Multivariate Analysis การวิเคราะห์มากกว่า 1 ตัวแปร ความสัมพันธ์ขั้นต้น
4. Detect Aberrant and Missing Values ข้อมูลที่ตกหล่นหายไป
5. Detect Outlier การวิเคราะห์หาข้อมูลที่ผิดปกติไปจากค่ากลาง หรือ Outlier
6. Feature Engineering การสร้าง Feature หรือตัวแปร


## ⚡ Install / Load Library

```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
```

## ⚡ Load ข้อมูล

```python
df1 = pd.read_csv("newPlayer15.csv")
```

## ⚡ EDA 5 ขั้นตอน

## ⚡ ⚓ 1. Distinguish Attributes  การเลือกข้อมูล และแสดงทำเป็น คอลัมภ์

```python
df2 = df1[['age','pace','shooting','passing','dribbling','defending','physic','skill_dribbling','skill_fk_accuracy','skill_long_passing','skill_ball_control','value_eur','league_level']]
```

```python
df2
```

 ⚓  ตรวจสอบ ประเภทข้อมูล ด้วย dtypes

 ⚓  ตรวจสอบ ภาพรวม ค่าว่าง ประเภท ด้วย info()

 ⚓  ตรวจสอบ ค่าสถิติเบื้องต้น ด้วย describe()

```python
# กำหนด scientific notation by setting float_format เพิ่มเติม https://pandas.pydata.org/docs/user_guide/options.html
pd.options.display.float_format = '{:.4f}'.format  # ทศนิยม 4 จุด

```

สัญกรณ์วิทยาศาสตร์ (อังกฤษ: scientific notation) คือรูปแบบของการเขียนตัวเลข ใช้โดยนักวิทยาศาสตร์ นักคณิตศาสตร์ หรือวิศวกร เพื่อให้สามารถเขียนหรือนำเสนอจำนวนที่มีขนาดใหญ่มากหรือเล็กมาก ให้ง่ายและสะดวกขึ้น

```python
df2.std()
```

 ⚓  ตรวจสอบ และ ลบ การซ้ำซ้อนข้อมูล ด้วย duplicated()

```python
# ตรวจสอบการซ้ำกันทั้งแถว
duplicate_rows_df = df2[df2.duplicated()]

print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# Dropping the duplicates
df2 = df2.drop_duplicates()
```

```python
df2.describe()
```

 ⚓  ตรวจสอบ การกระจาย ด้วย hist()

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()

df2.hist(ax = ax)
```

## ⚡ ⚓ 4. Detect Aberrant and Missing Values
การมองหาสิ่งผิดปกติในชุดข้อมูล และการมองหาข้อมูลที่ตกหล่นหายไป

## ⚓ นับจำนวนแถว (ที่มีข้อมูล) ในแต่ละคอลัมน์ เพื่อสังเกตุความผิดปกติขั้นต้น ด้วย count()

```python
df2.count()
```

## ⚓ สังเกตุความผิดปกติ สังเกตุตัวเลข ค่าว่าง ด้วย isnull() และ sum()

```python
# Finding the null values.

print(df2.isnull().sum())
```

## ⚓ ลบค่าว่าง dropna()

```python
# ลบค่าว่าง ด้วย Dropna()

df3 = df2.dropna()
df3
```

```python
print(df3.isnull().sum())
# # สังเกตุตัวเลข ค่าว่าง
```

## ⚡ ⚓ 3. Bi-/Multivariate Analysis
การวิเคราะห์มากกว่า 1 ตัวแปร เพื่อให้เห็นถึงความสัมพันธ์ขั้นต้น เช่น การหา Correlation และการเขียนกราฟ ​Scatterplot

```python
df3
```

 ⚓  ตรวจสอบ Correlation ด้วย corr()

```python
df3.corr()
```

 ⚓  ตรวจสอบ Correlation ด้วย corr() และ heatmap()

```python
# Finding the relations between the variables.
plt.figure(figsize=(20,10))
c= df3.corr()
sns.heatmap(c,cmap='BrBG',annot=True)
c
```

 ⚓  สำรวจความสัมพันธ์ แต่ละคอลัมน์ เป็นรายคู่ ด้วย Scatterplot

```python
# ค่าความสัมพันธ์ age vs pace
df3.age.corr(df3.pace)
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df3['age'], df3['pace'])
ax.set_xlabel('Age')
ax.set_ylabel('Pace')
plt.show()
```
