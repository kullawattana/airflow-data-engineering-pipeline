## 🧩 3.4 การใช้งานร่วมกับ Pandas

```python
# ใช้ Pandas

dfCountry = db["Country"]
df_Country = pd.DataFrame(list(dfCountry.find()))

df_Country
```

```python
dfState = db["states"]
df_State = pd.DataFrame(list(dfState.find()))

df_State
```

```python
# เลือกข้อมูลใน DataFrame แบบมีเงื่อนไข

US_df = df_State[df_State['country_code'] == 'US']

US_df
```

```python
# เลือกข้อมูลใน DataFrame แบบมีเงื่อนไข

dfusa = db["usa"]
df_usa = pd.DataFrame(list(dfusa.find()))

df_usa
```

```python
df_usa[df_usa['zip'] == 90003]
```

```python
# เลือกข้อมูลใน DataFrame แบบมีเงื่อนไข

df_usa_CA = df_usa[df_usa['state_id'] == 'CA']

df_usa_CA
```
